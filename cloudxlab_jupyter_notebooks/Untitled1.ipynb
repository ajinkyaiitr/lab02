```python
f = len(features)-1
M = np.random.randn(f,1)
```

```python
###Functions to be used
#1.line
#2.Error
#3.gradient_m
#4.gradient_c
#####
###Line Function
def line(m,x,c):
    return np.dot(x,m) + c
print(line(M,X,10).shape)
def error(m,x,c,y_a):
    y_p = line(m,x,c)
    return np.mean((y_p - y_a) **2)
print(error(M,X,10,y_a).shape)
```
